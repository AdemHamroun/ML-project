```python
df = pd.read_csv("Turbine_Data.csv",
                 low_memory=False,
                 parse_dates=["Unnamed: 0"])
```

```python
df['DateTime'] = df['Unnamed: 0']
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df['DateTime'].head(20)
```

```python
df['DateTime'] = pd.to_datetime(df['DateTime'],
 format = '%Y-%m-%dT%H:%M:%SZ',
 errors = 'coerce')

df['year'] = df['DateTime'].dt.year
df['month'] = df['DateTime'].dt.month
df['day'] = df['DateTime'].dt.day
df['hour'] = df['DateTime'].dt.hour
df['minute'] = df['DateTime'].dt.minute
```

```python
df[['year', 'month', 'day', 'hour', 'minute']].head()
```

```python
df.drop('DateTime', axis=1, inplace= True)
```

```python
df.isna().sum()
```

```python
df.isna().sum()
```

```python
for label, content in df.items():
    if pd.api.types.is_numeric_dtype(content):
        if pd.isnull(content).sum():
            # Add a binary column which tells if the data was missing our not
            df[label+"_is_missing"] = pd.isnull(content)
            # Fill missing numeric values with median since it's more robust than the mean
            df[label] = content.fillna(content.median())
```

```python
df.isna().sum()
```

```python
import seaborn as sns
plt.figure(figsize=(12, 6))

# Generate the heatmap
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('Figure 2.3 – Heatmap of Missing Values in the Dataset( after filling)')
plt.xlabel('Features')
plt.tight_layout()

# Save or show the figure
plt.savefig("missing_data_heatmap.png")  # Saves the image to file
plt.show()  # Displa
```

```python
df_with_time = df.copy()
df_with_time['DateTime'] = pd.to_datetime(df['year'].astype(str) + '-' + df['month'].astype(str) + '-' + df['day'].astype(str) + ' ' + df['hour'].astype(str) + ':' + df['minute'].astype(str))

df_with_time.sort_values("DateTime", inplace=True)

plt.figure(figsize=(15, 5))
plt.plot(df_with_time["DateTime"], df_with_time["ActivePower"], color='mediumseagreen')
plt.title("Active Power Over Time")
plt.xlabel("DateTime")
plt.ylabel("Active Power")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Sort by time just in case
df = df.sort_values(by=["year", "month", "day", "hour", "minute"])

# Split into train (before 2020) and test (2020 only)
train_df = df[df["year"] < 2020]
test_df = df[df["year"] == 2020]

# Separate features and target
X_train = train_df.drop("ActivePower", axis=1)
y_train = train_df["ActivePower"]

X_test = test_df.drop("ActivePower", axis=1)
y_test = test_df["ActivePower"]

# Check shapes
print(f"Train set: {X_train.shape}, Test set: {X_test.shape}")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
df
```

```python
categorical_columns = X_train.select_dtypes(include=['object']).columns
print("Categorical columns:", categorical_columns)
```

```python
# Apply One-Hot Encoding to the 'WTG' column
X_train_encoded = pd.get_dummies(X_train, columns=['WTG'])
X_test_encoded = pd.get_dummies(X_test, columns=['WTG'])

# Ensure the same columns in train and test sets (in case of differing categories)
X_train_encoded, X_test_encoded = X_train_encoded.align(X_test_encoded, join='left', axis=1, fill_value=0)

# Check the result
print(X_train_encoded.head())
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_encoded)
X_test_scaled = scaler.transform(X_test_encoded)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Initialize the model
model = LinearRegression()

# Train the model
model.fit(X_train_scaled, y_train)

# Predict on the test set
y_pred = model.predict(X_test_scaled)

# Evaluate the model
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f"Mean Absolute Error: {mae}")
print(f"Root Mean Squared Error: {rmse}")
```

```python
from sklearn.metrics import r2_score

# Calculate R² (R-squared) score
r2 = r2_score(y_test, y_pred)

print(f"R² Score: {r2}")
```

```python
import matplotlib.pyplot as plt

# Scatter plot of actual vs predicted values
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='blue', alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--')  # Line y = x (ideal prediction)
plt.title("Actual vs Predicted ActivePower")
plt.xlabel("Actual ActivePower")
plt.ylabel("Predicted ActivePower")
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Initialize the model
dt_model = DecisionTreeRegressor(random_state=42)

# Train the model
dt_model.fit(X_train_scaled, y_train)

# Predict on the test set
y_pred_dt = dt_model.predict(X_test_scaled)

# Evaluate the model
mae_dt = mean_absolute_error(y_test, y_pred_dt)
rmse_dt = np.sqrt(mean_squared_error(y_test, y_pred_dt))
r2_dt = r2_score(y_test, y_pred_dt)

print(f"Decision Tree Regressor:")
print(f"Mean Absolute Error (MAE): {mae_dt}")
print(f"Root Mean Squared Error (RMSE): {rmse_dt}")
print(f"R² Score: {r2_dt}")
```

```python
from sklearn.ensemble import RandomForestRegressor

# Initialize the model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Train the model
rf_model.fit(X_train_scaled, y_train)

# Predict on the test set
y_pred_rf = rf_model.predict(X_test_scaled)

# Evaluate the model
mae_rf = mean_absolute_error(y_test, y_pred_rf)
rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))
r2_rf = r2_score(y_test, y_pred_rf)

print(f"Random Forest - Mean Absolute Error: {mae_rf}")
print(f"Random Forest - Root Mean Squared Error: {rmse_rf}")
print(f"Random Forest - R² Score: {r2_rf}")

# Optional: Normalized errors
normalized_mae_rf = (mae_rf / y_test.max()) * 100
normalized_rmse_rf = (rmse_rf / y_test.max()) * 100

print(f"Random Forest - Normalized MAE: {normalized_mae_rf:.2f}%")
print(f"Random Forest - Normalized RMSE: {normalized_rmse_rf:.2f}%")

# Scatter plot: Actual vs Predicted
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred_rf, color='orange', alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--')
plt.title("Actual vs Predicted Active Power (Random Forest)")
plt.xlabel("Actual Active Power")
plt.ylabel("Predicted Active Power")
plt.show()
```

```python
# 1. Import necessary libraries
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt
import numpy as np

# 2. Initialize and train the model
xgb_model = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=6, random_state=42)
xgb_model.fit(X_train_scaled, y_train)

# 3. Predict on the test set
y_pred_xgb = xgb_model.predict(X_test_scaled)

# 4. Evaluate the model
mae_xgb = mean_absolute_error(y_test, y_pred_xgb)
rmse_xgb = np.sqrt(mean_squared_error(y_test, y_pred_xgb))
r2_xgb = r2_score(y_test, y_pred_xgb)

# 5. Print evaluation metrics
print(f"XGBoost - Mean Absolute Error: {mae_xgb}")
print(f"XGBoost - Root Mean Squared Error: {rmse_xgb}")
print(f"XGBoost - R² Score: {r2_xgb}")

# 6. Visualization: Actual vs Predicted
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred_xgb, color='orange', alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red', linestyle='--')
plt.title("Actual vs Predicted Active Power (XGBoost)")
plt.xlabel("Actual Active Power")
plt.ylabel("Predicted Active Power")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import matplotlib.pyplot as plt

# Assuming your data is already preprocessed:
# X_train_encoded, X_test_encoded, y_train, y_test are available

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_encoded)
X_test_scaled = scaler.transform(X_test_encoded)

# Define the MLP model
model = Sequential([
    Dense(128, activation='relu', input_shape=(X_train_scaled.shape[1],)),
    Dropout(0.2),
    Dense(64, activation='relu'),
    Dropout(0.2),
    Dense(1)  # Output layer for regression
])

# Compile the model
model.compile(optimizer='adam', loss='mse')

# Early stopping to avoid overfitting
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Fit the model
history = model.fit(
    X_train_scaled, y_train,
    validation_split=0.2,
    epochs=100,
    batch_size=256,
    callbacks=[early_stopping],
    verbose=1
)

# Make predictions
y_pred = model.predict(X_test_scaled).flatten()

# Evaluate performance
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"MLP - Mean Absolute Error: {mae}")
print(f"MLP - Root Mean Squared Error: {rmse}")
print(f"MLP - R² Score: {r2}")

# Visualization: Actual vs Predicted
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.5, color='purple')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')
plt.xlabel("Actual ActivePower")
plt.ylabel("Predicted ActivePower")
plt.title("MLP Model - Actual vs Predicted ActivePower")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot training & validation loss values
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss (MSE)')
plt.legend()
plt.grid(True)
plt.show()
```
